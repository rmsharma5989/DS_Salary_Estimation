from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS <br></br>
from nltk.corpuc import stopwords <br></br>
form nltk. tokenize import word_tokenize <br></br>

words = " ".join(df['Job Description']) <br></br>

def punctuation_stop(text): <br></br>
&nbsp;&nbsp;&nbsp;&nbsp; """ remove punctuation and stop word """ <br></br>
&nbsp;&nbsp;&nbsp;&nbsp; filtered = [] <br></br>
&nbsp;&nbsp;&nbsp;&nbsp; stop_words = set(stopwords.words('english')) <br></br>
&nbsp;&nbsp;&nbsp;&nbsp; word_tokens = word_tokenize(text) <br></br>
&nbsp;&nbsp;&nbsp;&nbsp; for w in word_tokens: <br></br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;if w not in stop_words and w.isalpha() <br></br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;filtered.append(w.lower()) <br></br>
&nbsp;&nbsp;&nbsp;&nbsp; return filtered <br></br>
 <br></br>

word_filetered = punctuation_stop(words) <br></br>
text = " ".join([ele for ele in word_filtered]) <br></br>
wc = WordCloud(background_color="white", random_state=1, stopwords=STOPWORDS, max_words=2000, width=800, height=1500) <br></br>
wc.generate(text) <br></br>

plt.figure(figsize=[10,10]) <br></br>
plt.imshow(wc, interpolation="bilinear") <br></br>
plt.axis('off') <br></br>
