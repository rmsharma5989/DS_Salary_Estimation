this above output is all statistical output of the data, it provide all statistical terms, there are very useful so lets understand it, but keep in mind
Data science is more focused on predictions and having models with predict on real words data, it is not as much concerned about correct specification of statistical problems.

**Dep. Variable:** this is target variable for which we are predicting the models, generally it is on Y axis  
**Model:** this is model name we used in this regression problem.  
**Method:** Method which was used to achieving the coefficient estimates, Least squares is the standard method (method used to fit the data into model)  
**No. Observations:** number of  observarions  
**Df Residuals:** the degree of freedom(DF) of the residuals, calculated by thaking the numbr of observations less than the number of parameters   
**Df Model:** the number of estimated parameters in the model, (Degree of freedom of the model)

**No. Observations = Df Residuals + Df Model + 1 (constant always left)**


**R-squared:** this is the coefficient of derermination, it measures the goodness of fit. (it basically tells the how much your data is fitted in the model)  

$$R^2 = 1 - \frac{SS_{res}}{SS_{tot}} $$  

**Adj. R-squared:** this is the adjusted R-squared. it is coefficient of determination adjusted by sample size and the number of parameter used. Adj R-squared is generally used in multiple regression, not in linear regression
$$ \bar{R}^2 = 1 - (1-R^2)\frac{n-1}{n-p-1} $$

> p = the total number of explanatory variables  
> n = the sample size

**F-statistic:** A measure that tells you if your model is differnt form simple average  

**Prob (F-statistic):** this measures the significance of your F-statistic, also called p-value of F-statistic, in statistics p-value equal or lower than 0.05 is considered significant.  

**AIC:** This is the Akike information criterion, it evaluates the model based on model complexity and number of observation, lower is better.  

**BIC:** This is Bayesian information criterion, similar to AIC, except it pushesh the model with more parameters

### Parameters  Estimated and the Associated Statistical Tests
**coef:** : The estimated coefficient test, note that this is just a point estiamtes  

**std err:** standard error if the estimate of the coefficient, Another term of standard deviation  

**t:** the t-statistic score  

**p>|t|:** The p-value, A measure of the probability that the coefficient is different from zreo. p-value greater then 0.05 if insignificant.  

**[95.0% Conf. Interval]:** The 95% confidience interval of the coefficient, Shown here as [0.025, 0.975]
