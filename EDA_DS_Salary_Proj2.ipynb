```python
#  Job simplifier
def data_simplifier(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower() or 'analytics' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'MLE'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    else:
        return 'NA'
    
# seniority
def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
        return 'Sr.'
    elif 'jr' in title.lower() or 'junior' in title.lower() or 'developer' in title.lower():
        return 'Jr'
    else:
        return 'NA'
```

```python
df.columns
```

```python
df["data_simp"] = df["Job Title"].apply(data_simplifier)
df.data_simp.value_counts()
```

```python
df["seniority"] = df["Job Title"].apply(seniority)
df.seniority.value_counts()
```

```python
# hourly wage to annual
pd.set_option("display.max_rows", None, "display.max_columns", None)
#df.apply(lambda x : x["Salary_type"].str != 'Glassdoor' and x["Salary_type"].str != 'Employer')
df[df["Salary_type"].str.contains('Per_hour')] [['Job Title', 'Salary Estimate', 'Job Description', 'Rating',
       'Company Name', 'Location', 'Headquarters', 'Size', 'Founded',
       'Type of ownership', 'Industry', 'Sector', 'Revenue', 'Competitors',
       'Salary_type', 'min_salary', 'max_salary', 'avg_salary', 'Company_text',
       'job_city', 'job_state', 'same_state', 'company_age', 'python_JD',
       'rstudio_JD', 'spark_JD', 'aws_JD', 'machinelearning_JD',
       'datascience_JD', 'comp_count']]
```

```python
# hourly wage to annual
# hour in a year = 22*8*12 = 2112, so we need to multiply hourly salary by 2112 and then devide by 1000 to get the K's -> 
#hence multiplay hourly salary by 2112/1000 = 2.112

df["min_salary"] = df.apply(lambda x : int(x.min_salary)*2.112 if x.Salary_type == 'Per_hour' else x.min_salary, axis = 1)
```

```python
df["max_salary"] = df.apply(lambda x : int(x.max_salary)*2.112 if x.Salary_type == 'Per_hour' else x.max_salary, axis = 1)
```

```python
df["avg_salary"] = df.apply(lambda x : (x.min_salary + x.max_salary)/2 if x.Salary_type == 'Per_hour' else x.avg_salary, axis=1)
```

```python
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import numpy as np
```

```python
set(stopwords.words('english'))
```

```python
words = " ".join(df['Job Description'])

filtered = []

stop_words = set(stopwords.words('english'))
word_tokens = word_tokenize(words)
for w in word_tokens:
    if w not in stop_words and w.isalpha():
        filtered.append(w.lower())

words_filtered = filtered

text = " ".join([ele for ele in words_filtered])

wc= WordCloud(background_color="white", random_state=1,stopwords=STOPWORDS, max_words = 2000, width =800, height = 1500)
wordCloud_img = wc.generate(text)

plt.figure(figsize=[20,20])
plt.imshow(wordCloud_img,interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
# a simple example of stop words
text = 'Hi Julia my name is John and i would like to talk to you for this tonight.'

from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

#text = "".join([w for w in text if w != 'unwanted']) # no use here but written to understand the functionalisy of Join and  
                                                    #  list comprehensans

stop_words = set(stopwords.words('english'))
words_tokens = word_tokenize(text)

filtered = []

for w in words_tokens:
    if w not in stop_words:
        filtered.append(w)

text = " ".join([w for w in filtered])

wc= WordCloud(background_color="white", random_state=1,stopwords=STOPWORDS, max_words = 2000, width =800, height = 1500)
wordCloud_img = wc.generate(text)

plt.figure(figsize=[7,7])
plt.imshow(wordCloud_img, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
text = '''In ancient manuscripts, another means to divide sentences into paragraphs was a line break (newline) followed by an
initial at the beginning of the next paragraph. An initial is an oversized capital letter, sometimes outdented beyond the
margin of the text. This style can be seen, for example, in the original Old English manuscript of Beowulf. Outdenting is
still used in English typography, though not commonly.[3] Modern English typography usually indicates a new paragraph by
indenting the first line. This style can be seen in the (handwritten) United States Constitution from 1787. For additional
ornamentation, a hedera leaf or other symbol can be added to the inter-paragraph white space, or put in the indentation
space.'''

from wordcloud import WordCloud,ImageColorGenerator, STOPWORDS
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stopword_text = set(stopwords.words('english'))

text = "".join([w for w in text])

tokenize_words = word_tokenize(text)

filtered = []

for w in tokenize_words:
    if w not in stopword_text:
        filtered.append(w)
        
text = " ".join([w for w in filtered])


wc = WordCloud(background_color ="white",height = 600, width = 900,stopwords=STOPWORDS,max_words=2000, random_state=1)    
img = wc.generate(text)

plt.figure(figsize=[10,10])
plt.imshow(img,interpolation='bilinear')
plt.axis("off")
plt.show()
```
